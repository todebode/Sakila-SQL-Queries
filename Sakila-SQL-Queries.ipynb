<b>1. What query would you run to get all the customers inside city_id = 312? Your query should return the customers' first name, last name, email, address, and city. 



```python
q = """SELECT customer.first_name, customer.last_name, customer.email, 
address.address, city.city
FROM customer
JOIN address ON customer.address_id = address.address_id
JOIN city ON address.city_id = city.city_id
WHERE city.city_id=312;"""
pd.read_sql(q, engine)
```

<b>2. What query would you run to get all comedy films? Note that the genre is called the category in this schema.  Your query should return film title, description, release year, rating, and special features.

```python
q = """SELECT film.title, film.description, film.release_year, film.rating, film.special_features
FROM film
JOIN film_category ON film_category.film_id = film.film_id
JOIN category ON category.category_id = film_category.category_id
WHERE category.name = 'Comedy';"""
pd.read_sql(q, engine)
```

<b>3. What query would you run to get all the films that Johnny Lollobrigida was in? Your query should return the actor's last name, film title, and release year.

```python
q = """SELECT (SELECT actor.last_name FROM actor 
WHERE actor.last_name = 'Lollobrigida' and actor.first_name = 'Johnny') 
AS Actor_LastName, 
film.title, film.release_year
FROM film;"""
pd.read_sql(q, engine)
```

<b>4. What query would you run to get the first and last names of all the actors in the movie titled "Bingo Talented"?

```python
q = """SELECT actor.first_name, actor.last_name, (SELECT film.title FROM film 
WHERE film.title = "Bingo Talented") AS film_title
FROM actor;"""
pd.read_sql(q, engine)
```

<b>5. What query would you run to get the customer_id associated with all payments greater than twice the average payment amount? (HINT: use 2* in your query to get twice the amount). Your result should include the customer id and the amount.

```python
q = """SELECT customer_id, amount
FROM payment
WHERE amount > 
  2*(SELECT AVG(amount)
  FROM payment);"""
pd.read_sql(q, engine)
```

<b>6. What query would you run to list the first and last names of the 5 customers who have the highest number(count) of payments? You can title the number of payments as num_payments.

```python
q = """SELECT customer.first_name, customer.last_name, 
	(SELECT COUNT(amount) FROM payment WHERE 
    payment.customer_id = customer.customer_id) AS num_payments
FROM customer
ORDER by num_payments DESC
LIMIT 5;"""
pd.read_sql(q, engine)
```
